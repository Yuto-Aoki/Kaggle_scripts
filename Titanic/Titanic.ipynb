```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import csv
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv("test.csv", header=0)
```

```python
train_df.head(20)
```

```python
test_df.isnull().sum()
```

```python
train_df['Sex'] = train_df['Sex'].map({'female': 0, 'male': 1}).astype(int)
dummy_df = pd.get_dummies(train_df['Embarked'])
train_df = pd.concat([train_df.drop(['Embarked'],axis=1),dummy_df],axis=1)
train_df.head()
```

```python
median_age = train_df['Age'].dropna().median()
train_df.loc[(train_df.Age.isnull()), "Age"] = median_age
```

```python
train_df = train_df.drop(["Name", "Ticket", "SibSp", "Cabin", "PassengerId"], axis=1)
```

```python
test_df["Sex"] = test_df["Sex"].map({"female": 0, "male": 1}).astype(int)
```

```python
median_age = test_df["Age"].dropna().median()
test_df.loc[(test_df.Age.isnull()), "Age"] = median_age
```

```python
dummy_df = pd.get_dummies(test_df['Embarked'])
test_df = pd.concat([test_df.drop(['Embarked'],axis=1),dummy_df],axis=1)
```

```python
ids = test_df["PassengerId"].values
```

```python
ids
```

```python
test_df = test_df.drop(["Name", "Ticket", "SibSp", "Cabin", "PassengerId"], axis=1)
```

```python
train_data = train_df.values.astype(int)
test_data = test_df.values.astype(int)
model = RandomForestClassifier(n_estimators=2000,max_depth=6,random_state=0, n_jobs=-1)
model.fit(train_data[0::, 1::], train_data[0::, 0])
output = model.predict(test_data).astype(int)
```

```python
submit_file = open("titanic_submit.csv", "w")
file_object = csv.writer(submit_file)
file_object.writerow(["PassengerId", "Survived"])
file_object.writerows(zip(ids, output))
submit_file.close()
```
